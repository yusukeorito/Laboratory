## Plaquete模型の相転移を見る

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
from itertools import cycle
plt.style.use("ggplot")
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
data_10 = np.loadtxt('../output/phase_transfer_L10_2025.dat')
data_20 = np.loadtxt('../output/phase_transfer_L20_2025.dat')
data_30 = np.loadtxt('../output/phase_transfer_L30_2025.dat')
data_40 = np.loadtxt('../output/phase_transfer_L40_2025.dat')
data_50 = np.loadtxt('../output/phase_transfer_L50_2025.dat')
```

```python
mag_10 = data_10[:,0]
mag_20 = data_20[:,0]
mag_30 = data_30[:,0]
mag_40 = data_40[:,0]
mag_50 = data_50[:,0]
T_10 = data_10[:, 1]
T_20 = data_20[:, 1]
T_30 = data_30[:, 1]
T_40 = data_40[:, 1]
T_50 = data_50[:, 1]
```

```python
idx = np.argwhere(np.sign(mag_10 - mag_20) == 0)
print(idx)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
l1, l2, l3, l4, l5 = "L=10", "L=20","L=30","L=40", "L=50" 
ax.plot(T_10, mag_10, label=l1,marker='o', markersize=5, linestyle='None')
ax.plot(T_20, mag_20, label=l2,marker='o', markersize=5, linestyle='None')
ax.plot(T_30, mag_30, label=l3,marker='o', markersize=5, linestyle='None')
ax.plot(T_40, mag_40, label=l4,marker='o', markersize=5, linestyle='None')
ax.plot(T_50, mag_50, label=l5,marker='o', markersize=5, linestyle='None')
#ax.set_xlabel('T', fontsize=16)
#ax.set_ylabel('abs(M/N)', fontsize=16)
ax.legend(fontsize=20)
ax.set_xticks(np.linspace(1,6,11))
#plt.suptitle('Phase Transition', fontsize=16)
plt.tight_layout()
plt.show()
#plt.plot(T, mag, marker='o', markersize=5, linestyle='None')
```
